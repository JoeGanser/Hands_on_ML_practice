```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
#mnist
```

```python
import numpy as np
import pandas as pd
X, y = mnist['data'], mnist['target']
#sample = pd.DataFrame(mnist).sample(2000)
#Xs,ys = sample['data'],sample['target']
print('X shape: ',X.shape)
print('Y shape: ',y.shape)
```

```python
X_train = X[:60000]
X_test = X[60000:]
y_train = y[:60000]
y_test = y[60000:]
```

```python
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score
rf = RandomForestClassifier()

def model_time(X_train,X_test,y_train,y_test):
    t0=time.time()
    rf.fit(X_train, y_train)
    train_time = time.time()-t0
    t1 = time.time()
    pred = rf.predict(X_test)
    predict_time = time.time()-t1
    Time_total = train_time+predict_time
    
    acc = accuracy_score(y_test, pred)
    return Time_total,acc
    
time_without_pca,acc_without_pca = model_time(X_train,X_test,y_train,y_test)
print('Time without pca: ',time_without_pca)
print('accuracy without pca: ',acc_without_pca)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)

X_p_train = X_reduced[:60000]
X_p_test = X_reduced[60000:]
```

```python
time_with_pca,acc_with_pca = model_time(X_p_train,X_p_test,y_train,y_test)

print('Time with 95% PCA: ',time_with_pca)
print('accuracy with 95% PCA: ',acc_with_pca)
```

```python
from sklearn import datasets, svm, metrics

# The digits dataset
digits = datasets.load_digits()
```

```python
X = pd.DataFrame(digits['data'])
y = pd.DataFrame(digits['target'])
```

```python
data = pd.concat([pd.DataFrame(tsne),pd.DataFrame(y)],axis=1)
data.columns = [0,1,3]
```

```python
import matplotlib
import matplotlib.pyplot as plt

colors = ['red','green','blue','purple','orange',
          'yellow','black','violet','grey','brown']

label = data[3]
plt.scatter(data[0],data[1],c=label, cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df2 = pd.DataFrame(pca.fit_transform(X))
df2 = pd.concat([df2,y],axis=1)
df2.columns=[0,1,2]
df2.head()
```

```python
colors = ['red','green','blue','purple','orange',
          'yellow','black','violet','grey','brown']

label = df2[2]
plt.scatter(df2[0],df2[1],c=label, cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```

```python

from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10, random_state=42)
X_lle = pd.DataFrame(lle.fit_transform(X))

X_lle = pd.concat([X_lle,y],axis=1)
X_lle.columns = [0,1,2]


plt.scatter(X_lle[0],X_lle[1],c=X_lle[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```

```python

from sklearn.manifold import MDS

mds = MDS(n_components=2, random_state=42)
X_mds = pd.DataFrame(mds.fit_transform(X))

X_mds = pd.concat([X_mds,y],axis=1)
X_mds.columns =[0,1,2]

plt.scatter(X_mds[0],X_mds[1],c=X_mds[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.show()
```
