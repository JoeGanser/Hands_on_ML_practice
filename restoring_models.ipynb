# Reusing pre-trained layers

First you need to load the graph's structure. The `import_meta_graph()` function does just that, loading the graph's operation into the default graph and returning a saver that you can then use to restore the model's state. Note that by default, a Saver saves the structure into a graph using the meta file, so thats the file you should load.

```python
import tensorflow as tf
tf.reset_default_graph()
```

```python
saver = tf.train.import_meta_graph("./tmp2/multi_classifier_model_final.ckpt.meta")
```

```python
for op in tf.get_default_graph().get_operations():
    print(op.name)
```
